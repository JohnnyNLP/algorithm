```python
def solution(s):
    answer = True
    stack = []

    # 개수 비교
    if s.count(')') != s.count('('):
        return False    
    
    # 시작과 끝 먼저 점검
    if s.startswith(')') or s.endswith('('):
        return False
    
    # 쉬운 쌍 전처리
    while '()' in s:
        s = s.replace('()', '')
        
    if len(s) == 0:
        return True
    
    for i in range(len(s)):
        stack.append(s[i])
        
        # 스택의 가장 아래에 '('가 있고, 가장 위에 ')'가 있는 상황마다 이를 pop해준다.
        while len(stack) >= 2 and (stack[0]=='(' and stack[-1]==')'):
            stack.pop(-1)
            stack.pop(0)
            print(stack)
            
            # 스택의 마지막에 ')'가 쌓여 있는 상황이라면 즉시 종료한다.
            if stack and stack[0] == ')':
                return False
            
        
        
    # stack이 비어 있다면 true, 아니면 false
    if stack:
        return False
    else:
        return True
```
